```python
def rank_performance (stock_price):
    if stock_price >= 950:
        return "Excellent"
    elif stock_price <=400:
        return "Poor"
    else:
        return "Decent"
```

```python
plt.style.use("ggplot")
shop['Close'].apply(rank_performance).value_counts().plot(kind = "barh")
```

```python
shop['Close'].apply(rank_performance).value_counts().plot(kind = "pie")
```

```python
def rank_performance(stock_price):
    if stock_price > shop["Close"].mean():
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("dark_background")
shop["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True)
```
